```python
import pandas as pd

from litecoder.us_cities import USCityIndex
from litecoder.us_states import USStateIndex
```

```python
city_idx = USCityIndex()
city_idx.build()
```

```python
state_idx = USStateIndex()
state_idx.build()
```

```python
city_idx
```

```python
state_idx
```

```python
locs = pd.read_json('/Users/dclure/Projects/geovec-data/data/locations-90k.json', lines=True)
```

```python
for r in locs.dropna().head(10000).itertuples():
    
    res = []
    
    res += city_idx.query(r.location)
    res += state_idx.query(r.location)
    
    if res:
        print('%s ~ %s' % (r.location, res))
```

```python
total = 0
for r in locs.dropna().itertuples():
    
    res = []
    
    res += city_idx.query(r.location)
    res += state_idx.query(r.location)
    
    if res:
        total += r.count
```
